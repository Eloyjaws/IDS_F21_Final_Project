```python
pd.options.mode.chained_assignment = None  # default='warn'

edition_to_year = {
    'E1': 2017,
    'E2': 2018,
    'E3': 2019,
    'E4': 2020,
    'E5': 2021
}

def get_year(edition):
    return edition.apply(lambda edition: edition_to_year[edition])

df_3i_corrected['Year'] = get_year(df_3i_corrected['Edition'])

df_3i_corrected[['Country', 'Year', 'Edition', 'Internet users (percent of population)', 'Internet users (population)', 'Population', 'Offline population']].query("Country == 'United States'")

```

## 1. Share of population using the internet

```python
def get_top_k(df, sort_by, k=15, ascending=False):
    # Top K after grouping inspired by 
    # https://stackoverflow.com/a/66551484

    return df.groupby('Year').apply(lambda x : x.sort_values(by = sort_by, ascending = ascending).head(k)).reset_index(drop=True)

```

```python
df_3i_corrected.columns
```

```python


df_internet_users = df_3i_corrected[['Year', 'Country', 'Internet users (population)', 'Offline population', 'Internet users (percent of population)', 'Population', 'Mobile subscribers (Percentage)' ]]

df_internet_users_top_15 = get_top_k(df_internet_users, 'Internet users (population)')
df_internet_users_bottom_15 = get_top_k(df_internet_users, 'Internet users (population)', ascending=True)

df_internet_users_top_15.to_csv("df_internet_users_top_15.csv", index=False)
df_internet_users_bottom_15.to_csv("df_internet_users_bottom_15.csv", index=False)


```

```python
bars = (
    alt.Chart(df_internet_users_bottom_15.query('Year == 2021'))
    .mark_bar()
    .encode(
        alt.X('Country', sort=None),
        alt.Y('Internet users (population)', 
        sort=['Internet users (population)']),
        color='Country',
    )
)
plot_title = f"Bottom 15 countries ranked by Internet Users (in millions)"
bars.properties(title=plot_title)
```

```python
bars = (
    alt.Chart(df_internet_users_top_15.query('Year == 2021'))
    .mark_bar()
    .encode(
        alt.X('Country', sort=None),
        alt.Y('Internet users (population)', 
        sort=['Internet users (population)']),
        color='Country',
    )
)
plot_title = f"Top 15 countries ranked by Internet Users (in millions)"
bars.properties(title=plot_title)
```

```python


df_percentage_internet_users = df_3i_corrected[['Year', 'Country', 'Internet users (population)', 'Offline population', 'Internet users (percent of population)', 'Population', 'Mobile subscribers (Percentage)' ]]

df_percentage_internet_users_top_15 = get_top_k(df_percentage_internet_users, 'Internet users (percent of population)')
df_percentage_internet_users_bottom_15 = get_top_k(df_percentage_internet_users, 'Internet users (percent of population)', ascending=True)

df_percentage_internet_users_top_15.to_csv("df_percentage_internet_users_top_15.csv", index=False)
df_percentage_internet_users_bottom_15.to_csv("df_percentage_internet_users_bottom_15.csv", index=False)
```

```python
bars1 = (
    alt.Chart(df_percentage_internet_users_top_15.query('Year == 2021'))
    .mark_bar()
    .encode(
        alt.X('Country', sort=None),
        alt.Y('Internet users (percent of population)', 
        sort=['Internet users (percent of population)']),
        color='Country',
    )
)
plot_title = f"Top 15 countries ranked by Percentage of the population with Internet access"
bars1.properties(title=plot_title)

bars2 = (
    alt.Chart(df_percentage_internet_users_bottom_15.query('Year == 2021'))
    .mark_bar()
    .encode(
        x = alt.X('Country', sort=None),
        y = alt.Y('Internet users (percent of population)'),
        color='Country',
    )
)
plot_title = f"Bottom 15 countries ranked by Percentage of the population with Internet access"
bars2.properties(title=plot_title)


bars1 | bars2
alt.hconcat(bars1, bars2, title="Top (L) and Bottom (R) 15 countries ranked by Percentage of the population with Internet access")


```

```python


df_percentage_offline_population = df_3i_corrected[['Year', 'Country', 'Internet users (population)', 'Offline population', 'Offline population (Percentage)', 'Internet users (percent of population)', 'Population', 'Mobile subscribers (Percentage)' ]]

df_percentage_offline_population_top_15 = get_top_k(df_percentage_offline_population, 'Offline population (Percentage)')
df_percentage_offline_population_bottom_15 = get_top_k(df_percentage_offline_population, 'Offline population (Percentage)', ascending=True)

df_percentage_offline_population_top_15.to_csv("df_percentage_offline_population_top_15.csv", index=False)
df_percentage_offline_population_bottom_15.to_csv("df_percentage_offline_population_bottom_15.csv", index=False)
```

```python
bars1 = (
    alt.Chart(df_percentage_offline_population_top_15.query('Year == 2021'))
    .mark_bar()
    .encode(
        alt.X('Country', sort=None),
        alt.Y('Offline population (Percentage)'),
        color='Country',
    )
)
plot_title = f"Top 15 countries ranked by Percentage of the population without Internet access"
bars1.properties(title=plot_title)

bars2 = (
    alt.Chart(df_percentage_offline_population_bottom_15.query('Year == 2021'))
    .mark_bar()
    .encode(
        x = alt.X('Country', sort=None),
        y = alt.Y('Offline population (Percentage)'),
        color='Country',
    )
)
plot_title = f"Bottom 15 countries ranked by Percentage of the population without Internet access"
bars2.properties(title=plot_title)


bars1 | bars2
alt.hconcat(bars1, bars2, title="Top (L) and Bottom (R) 15 countries ranked by Percentage of the population with Internet access")


```
